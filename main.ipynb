```python
from dotenv import load_dotenv
import google.generativeai as genai
import pymongo
import os

# Load environment variables
load_dotenv()

# Set up Gemini API Key
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))

# MongoDB Connection
MONGO_URI = os.getenv("MONGO_URI")
client = pymongo.MongoClient(MONGO_URI)
db = client["mistakes"]
mistakes_collection = db["mistakes"]

def get_user_preferences():
    known_language = input("What language do you know? ")
    target_language = input("What language do you want to learn? ")
    proficiency = input("What is your proficiency level (beginner/intermediate/advanced)? ")
    return known_language, target_language, proficiency

def analyze_mistake(user_input, correction):
    mistakes_collection.insert_one({"user_input": user_input, "correction": correction})

def get_mistake_summary():
    mistakes = list(mistakes_collection.find({}, {"_id": 0}))
    if mistakes:
        print("\nSummary of mistakes:")
        for mistake in mistakes:
            print(f"You said: {mistake['user_input']}, Correction: {mistake['correction']}")
    else:
        print("No mistakes recorded.")

def chat_with_bot(known_language, target_language, proficiency):
    try:
        available_models = [model.name for model in genai.list_models()]
        print("Available models:", available_models)

        valid_models = [
            name for name in available_models
            if "generateContent" in genai.get_model(name).supported_generation_methods
            and "vision" not in name.lower()
            and not name.startswith("models/chat-bison")
        ]

        if not valid_models:
            print("No valid models that support content generation.")
            return

        preferred = "models/gemini-1.5-pro"
        model_name = preferred if preferred in valid_models else valid_models[0]
        print(f"Using model: {model_name}")

        chat_model = genai.GenerativeModel(model_name)

    except Exception as e:
        print("Error initializing model:", str(e))
        return

    system_message = f"You are a language tutor helping users learn {target_language}."
    print("\nStarting conversation. Type 'exit' to quit.")
    while True:
        user_input = input(f"[{target_language}] You: ")
        if user_input.lower() == "exit":
            break

        try:
            response = chat_model.generate_content([system_message, user_input])
            print(f"Bot: {response.text if response else 'No response received'}")

            # Simulate mistake tracking (you can improve this with NLP)
            if "wrong" in response.text.lower():
                analyze_mistake(user_input, response.text)
        except Exception as e:
            print("Error generating response:", str(e))

    get_mistake_summary()

if __name__ == "__main__":
    known_language, target_language, proficiency = get_user_preferences()
    chat_with_bot(known_language, target_language, proficiency)
```
